```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/drive/MyDrive/colab_img/histogram_equaliztion.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def show(image):
  plt.axis("off")
  plt.imshow(image, cmap="gray")
  plt.show()

equalization = cv2.equalizeHist(image)

show(image)
show(equalization)
```

```python
show(image)
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize=(8,8))
clahe_image = clahe.apply(image)
show(equalization)
show(clahe_image)
```
